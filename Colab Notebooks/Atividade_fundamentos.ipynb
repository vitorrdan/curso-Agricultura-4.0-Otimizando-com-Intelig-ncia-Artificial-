```python
zip_file_name = "/content/2024.zip"
import pandas as pd
import zipfile

with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    with zip_ref.open('INMET_S_SC_A895_CHAPECO_01-01-2024_A_30-06-2024.CSV') as csv_file:

        df = pd.read_csv(csv_file, sep=';',skiprows=8, encoding='latin-1',on_bad_lines='skip')
```

```python
for column in df:
  if df[column].dtype == 'object':
    df[column] = df[column].str.replace(',', '.')
    df[column] = pd.to_numeric(df[column], errors= 'coerce')
```

```python
def add_labels(df):
  labels = []
  for precip in df['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)']:
    if precip < 4:
      labels.append('0 a 4 mm')
    elif precip < 8:
      labels.append('4 a 8 mm')
    elif precip < 12:
      labels.append('8 a 12 mm')
    elif precip < 16:
      labels.append('12 a 16 mm')
    elif precip < 20:
      labels.append('16 a 20 mm')
    elif precip < 24:
      labels.append('20 a 24 mm')
    else:
      labels.append('24 a 28 mm')
  df['labels'] = labels

add_labels(df)
```

6. Faça a normalização e a padronização de df["VENTO, DIREÇÃO HORARIA (gr) (° (gr))"], e mostre os gráficos da distribuição antes e após procedimento os procedimentos. Podemos dizer que é uma distribuição normal ? Justifique.

```python
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Normalização
scaler = MinMaxScaler()
df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_norm'] = scaler.fit_transform(df[['VENTO, DIREÇÃO HORARIA (gr) (° (gr))']])

# Visualizando os dados originais e normalizados
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))'], bins=10, color='blue', alpha=0.7)
plt.title('Dados Originais')

plt.subplot(1, 2, 2)
plt.hist(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_norm'], bins=10, color='green', alpha=0.7)
plt.title('Dados Normalizados')

plt.tight_layout()
plt.show()
```

```python
# Padronização
scaler = StandardScaler()
df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_std'] = scaler.fit_transform(df[['VENTO, DIREÇÃO HORARIA (gr) (° (gr))']])

# Visualizando os dados originais e padronizados
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))'], kde=True, color='blue', bins=10, ax=axes[0])
axes[0].set_title('Distribuição Original')
axes[0].set_xlabel('Valores')
axes[0].set_ylabel('Densidade')

sns.histplot(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_std'], kde=True, color='green', bins=10, ax=axes[1])
axes[1].set_title('Distribuição Padronizada')
axes[1].set_xlabel('Valores Padronizados')
axes[1].set_ylabel('Densidade')

plt.tight_layout()
plt.show()

# Estatísticas
print('Dados Originais:')
print('Média:', np.mean(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))']))
print('Desvio padrão:', np.std(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))']))
print('Variância:', np.var(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))']))

print('\nDados Padronizados:')
print('Média:', np.mean(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_std']))
print('Desvio padrão:', np.std(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_std']))
print('Variância:', np.var(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_std']))
```

```python
#Analise de normalidade
from scipy.stats import shapiro, probplot

# Teste de Shapiro-Wilk para dados originais
shapiro_test_original = shapiro(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))'])
print(f"Shapiro-Wilk dados originais: {shapiro_test_original}")

# Gráfico Q-Q para dados originais
plt.figure(figsize=(6, 6))
probplot(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))'], dist="norm", plot=plt)
plt.title('Q-Q Plot dados originais')
plt.show()

# Teste de Shapiro-Wilk para dados padronizados
shapiro_test_std = shapiro(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_std'])
print(f"Shapiro-Wilk dados padronizados: {shapiro_test_std}")

# Gráfico Q-Q para dados padronizados
plt.figure(figsize=(6, 6))
probplot(df['VENTO, DIREÇÃO HORARIA (gr) (° (gr))_std'], dist="norm", plot=plt)
plt.title('Q-Q Plot dados padronizados')
plt.show()
```

Após visualizar os gráficos de distribuição e os gráficos Q-Q, junto com os resultados do teste de Shapiro-Wilk, podemos considerar que é uma distribuição normal tendo em vista o p-valor do teste de Shapiro-Wilk ser maior que 0.05 e os dados seguem aproximadamente uma linha reta no gráfico Q-Q
